```python
import seaborn
from matplotlib.pyplot import *

%matplotlib inline

import numpy as np
import cneurons as cn

from scipy import ndimage
y = np.loadtxt('20120406_1_3_15i.dat')
```

```python
e = cn.mat()
e.apply_current(y,0.02);
r, rspikes = e.simulate(2000,1)
r[rspikes,0] = 40
```

```python
seaborn.set_style("ticks")
figure(figsize=(9,3))
plot(r[:,0]*2 - 70,linewidth=2)
vlines(rspikes,35,40)
xlabel("ms")
ylabel('"voltage"')
print(rspikes)
seaborn.despine(trim=True,offset=5)
```
